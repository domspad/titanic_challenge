```python
import mechanize
from bs4 import BeautifulSoup
import pandas as pd
```

```python
fulldatafile = '../data/full.csv'
df = pd.read_csv(fulldatafile) #index_col='PassengerId')
df.info()
```

#### Scraping the Name and Survived of riders of titanic into a dataframe 'newdf'

```python
br = mechanize.Browser()
```

```python
br.set_handle_robots(False)
```

```python
br.addheaders = [("User-agent","Mozilla/5.0 (X11; U; Linux i686; en-US; rv:1.9.2.13) Gecko/20101206 Ubuntu/10.10 (maverick) Firefox/3.6.13")]
```

```python
soup = BeautifulSoup(br.open('http://www.encyclopedia-titanica.org/titanic-passengers-and-crew/'))
```

```python
ppl = soup.findAll('tr',{'id' :'infinite_scroll'})
```

```python
records = []
for p in ppl :
    attributes = p.findAll('td')
    newpass = {'Name' : attributes[0].text.replace(u'\n',u''),
               'Survived' : str(attributes[6].text.replace(u'\xa0',u''))}
    records.append(newpass)
```

```python
newdf = pd.DataFrame.from_records(records)
```

```python
cutoff = 16
```

```python
filldf['Last'] = filldf.Name.map(lambda x : x[:x.find(',')].lower())
filldf['First'] = filldf.Name.map(lambda x : x.replace('.','')[x.find(',')+1:x.find(',')+1+cutoff].lower())
```

```python

newdf['Last'] = newdf.Name.map(lambda x : x[:x.find(',')].lower())
newdf['First'] = newdf.Name.map(lambda x : x[x.find(',')+1:x.find(',')+1+cutoff].lower())
```

```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
from itertools import combinations_with_replacement

def polyfeat(df, collist, p=2) :
#     returns df with additional columns df[col1]*df[col2]*..*df[colp] for all
#     col1,col2,...,colp in the collist (with replacement)
    colflag = {col: False for col in collist}
    for col in collist :
        if len(np.unique(df[col].values)) < 3 : #so binary vars don't get
            colflag[col] = True# needlessly multiplied
    for cols in combinations_with_replacement(collist,p) :
        needed = True
        for col in cols :
            if colflag[col] and cols.count(col) > 1 :
                needed = False
        if not needed :
            continue
        name = '*'.join(cols)
        df[name] = 1
        for col in cols : 
            df[name] = df[name] * df[col]
```

### Data cleaning

```python
df['Gender'] = df.Sex.map({'male': 0,
                            'female' : 1})
```

```python
median_ages = np.zeros((2,3))

for i in range(0,2) :
    for j in range(0,3) :
        median_ages[i,j] = df[ (df['Gender'] == i) &
                               (df['Pclass'] == j+1)].Age.dropna().median()
```

```python
df['AgeFill'] = df['Age']
```

```python
for i in range(0,2):
    for j in range(0,3) :
        df.loc[ (df.Age.isnull()) &
               (df['Pclass'] == j+1) &
               (df['Gender'] == i),
               'AgeFill'] = median_ages[i,j]
```

```python
df['AgeisNull'] = df.Age.isnull().astype(int)
```

```python
portdf = pd.get_dummies(df.Embarked,prefix='port')
df = pd.merge(df,portdf,left_index=True,right_index=True)
```

#### dropping columns not to be used in next ML

```python
dropcols = ['Sex','Age','Name','Ticket','Cabin','Embarked']
df.drop(dropcols,axis=1,inplace=True)
```

## Feature Engineering

```python
df['FamilySize'] = df['SibSp'] + df['Parch']
```

```python
polycols = ['Pclass','Fare','Gender','AgeFill','FamilySize']
polyfeat(df, polycols, 2)
polyfeat(df, polycols, 3)
```

## Scaling (Same)

```python
mus = {}
sigs = {}
```

```python
for col in df.columns :
    if (min(df[col]) < -3) or (max(df[col]) > 3) :
        mus[col] = df[col].mean()
        sigs[col] = df[col].std()
```

```python
for col in mus :
    df[col] = (df[col] - mus[col])/(2 * sigs[col])
```

- AgeisNull, CabinisNull
    - Nans since less info generally means we don't know because they didn't survive
- AgeFill (with median from Pclass,gender combo)
    - As a way to deal with Nans in Age category (not sure if better than 0fill though)
- Age^2 or bucketed classes (did buckets)
    - Because age doesn't seem to have linear relationship with survival
- Fare^2 or bucketed classes (did buckets)
    - Because little seems to be explained by Fare alone (may not be linear relationship)
- FamilySize = sum of Sibsp and Parch
    - (suggested)... just because sounds logical
- Master = if 'Master' in title (didn't do b/c linearly dependent with sibsp and parch)
    - Many Masters misclassified in cv set
- Miss = if 'Miss' in title
    - fractions maybe be equally represented in wrongcv and original df but could be useful
- Mrs = if 'Mrs' in title
    - fractions maybe be equally represented in wrongcv and original df but could be useful
- Pclass as binary vector (for nonlinear relationship)
    - doesn't seem to be linear relationship (lot of 1's, 3's misclassified)
- CabinLetter
    - Some 'E's seem to be commonly misclassified
- Alone = if Parch and Sibsp 0
    - A lot of alone males wrongl classified as survived

### Data cleaning

```python
df['Gender'] = df.Sex.map({'male': 0,
                            'female' : 1})
```

```python
median_ages = np.zeros((2,3))

for i in range(0,2) :
    for j in range(0,3) :
        median_ages[i,j] = df[ (df['Gender'] == i) &
                               (df['Pclass'] == j+1)].Age.dropna().median()
```

```python
df['AgeFill'] = df['Age']
```

```python
for i in range(0,2):
    for j in range(0,3) :
        df.loc[ (df.Age.isnull()) &
               (df['Pclass'] == j+1) &
               (df['Gender'] == i),
               'AgeFill'] = median_ages[i,j]
```

```python
fig,ax = plt.subplots(nrows=1,ncols=2, sharey=True)

df.AgeFill.hist(alpha=0.5,ax=ax[0])
df.Age.hist(alpha=0.5,ax=ax[1])
plt.show()
```

```python
df['AgeisNull'] = df.Age.isnull().astype(int)
df['CabinisNull'] = df.Cabin.isnull().astype(int)
```

```python
df.Age.hist(bins=5)
```

```python
binsage = [  0.17,  16.336  ,  32.252  ,  48.168  ,  64.084  ,  80.     ]
binsfare = [0,8,15,25,40,55,100,512.3293] #train data fare max
```

```python
vals = pd.cut(df.AgeFill, bins=binsage)
df = pd.merge(df, pd.get_dummies(vals,prefix='AgeFill'),left_index=True,right_index=True)
```

```python
df.Fare.hist(bins=5, range=(0,16))
plt.show()
df.Fare.hist(bins=binsfare)
plt.show()
```

```python
vals = pd.cut(df.Fare,bins=binsfare)
df = pd.merge(df,pd.get_dummies(vals,prefix='Fare'),left_index=True,right_index=True)
```

```python
portdf = pd.get_dummies(df.Embarked,prefix='port')
df = pd.merge(df,portdf,left_index=True,right_index=True)
```

```python
df = pd.merge(df,pd.get_dummies(df.Pclass,prefix='Pclass'), left_index=True, right_index=True)
```

## Feature Engineering

```python
masters_idx = df.Name.apply(lambda x: 'Master' in x)
mr_idx = df.Name.apply(lambda x: 'Mr.' in x)
dr_idx = df.Name.apply(lambda x: 'Dr.' in x) # 7 doctors
rev_idx = df.Name.apply(lambda x: 'Rev.' in x)
masters = masters_idx.astype(int)
mr = mr_idx.astype(int)
dr = dr_idx.astype(int)
rev = rev_idx.astype(int)
df['Master'] = masters
df['Mr'] = mr
df['Dr'] = dr
# df['Dr'] = dr
df[~(mr_idx | masters_idx) & ~(df.Sex == 'female')] #special men!
```

```python
miss_idx = df.Name.apply(lambda x: 'Miss' in x)
mrs_idx = df.Name.apply(lambda x: 'Mrs' in x)
miss = miss_idx.astype(int)
mrs = mrs_idx.astype(int)
df['Miss'] = miss
df['Mrs'] = mrs
df[~(miss_idx | mrs_idx) & ~(df.Sex == 'male')] #special women!
```

```python
def other_cat(cabin) :
    if isinstance(cabin,str) :
        if cabin[0] in ['A','B','C','D','E','F'] :
            return cabin[0]
        else:
            return 'other'
    else :
        return cabin
```

```python
cabins = df.Cabin.apply(lambda x : other_cat(x))
df = pd.merge(df, pd.get_dummies(cabins,prefix='Cabin'), left_index=True, right_index=True)
```

```python
df['Alone'] = ((df.SibSp + df.Parch)==0).astype(int)
```

```python
dropcols = ['Pclass','Name','Sex','Age','Ticket','Cabin','Embarked']
df.drop(dropcols,axis=1,inplace=True)
```

## Scaling (Same)

```python
mus = {}
sigs = {}
```

```python
for col in df.columns :
    if (min(df[col]) < -3) or (max(df[col]) > 3) :
        mus[col] = df[col].mean()
        sigs[col] = df[col].std()
```

```python
for col in mus :
    df[col] = (df[col] - mus[col])/(2 * sigs[col])
```
